```python
import numpy as np
import pandas as pd
import re
import pickle
```

```python
with open('./with_name/LDA_clean_scaled_final.pkl', 'rb') as f:
    df_with = pickle.load(f)
```

```python
df_with
```

```python
df_with.drop(['문서 번호', '가장 비중이 높은 토픽', '가장 높은 토픽의 비중', '각 토픽의 비중', 'str_len'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split

train_with, test_with = train_test_split(df_with, test_size=.2)
```

```python
counter = 1
for _, line in train_with.iterrows():

    if line["주가변동"] == 0:
        sen = line["뉴스제목"]
        filename = './with_name/classes_f/train/0/{0:06d}.txt'.format(counter)
        text_file = open(filename, "w", encoding='utf-8')
        text_file.write(sen)
        text_file.close()
    else:
        sen = line["뉴스제목"]
        filename = './with_name/classes_f/train/1/{0:06d}.txt'.format(counter)
        text_file = open(filename, "w", encoding='utf-8')
        text_file.write(sen)
        text_file.close()

    counter += 1

counter = 1
for _, line in test_with.iterrows():

    if line["주가변동"] == 0:
        sen = line["뉴스제목"]
        filename = './with_name/classes_f/test/0/{0:06d}.txt'.format(counter)
        text_file = open(filename, "w", encoding='utf-8')
        text_file.write(sen)
        text_file.close()
    else:
        sen = line["뉴스제목"]
        filename = './with_name/classes_f/test/1/{0:06d}.txt'.format(counter)
        text_file = open(filename, "w", encoding='utf-8')
        text_file.write(sen)
        text_file.close()

    counter += 1
```
